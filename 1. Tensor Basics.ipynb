## **Convert a Numpy array (VECTOR) to Tensor(1D TENSOR)**

```python
# Create an Numpy Array

arr = np.array([1,2,3,4])
```

```python
arr
```

```python
arr.dtype
```

### **Method 1**

```python
# Convert Numpy Array to tensor

# Method 1

x = torch.from_numpy(arr)
```

```python
x
```

```python
x.dtype
```

### **Method 2**

```python
# Method 2

y = torch.as_tensor(arr)

y
```

```python
y.dtype
```

### **Method 3**

```python
z = torch.tensor(arr)
z
```

```python
z.dtype
```

### **Difference between above 3 Methods**

```python
# changing the value of 1st element of numpy array

# Before Change

arr
```

```python
# After Change

arr[0] = 999
arr
```

```python
# Method 1 Value
x
```

```python
# Method 2 Value
y
```

```python
# Method 3 Value
z
```

if I change the value of numpy array Mehtod 1 and Method 2 Tensor change its values, as it shared same memory location

Method 3 tensor won't change, as it copy the value and use different Memory Location

## **Convert a 2d Numpy array (MATRIX) to Tensor(2D TENSOR)**

```python
arr2d = np.arange(0.0, 12.0)
arr2d
```

```python
# Reshaping the array to 2D Matric

arr2d = arr2d.reshape(3,4)
arr2d
```

```python
x2 = torch.from_numpy(arr2d)
x2
```

```python
x2.dtype
```

```python
# Contain Integer Values

torch.tensor(arr)
```

```python
# Contain Floating Values

torch.Tensor(arr)
```
