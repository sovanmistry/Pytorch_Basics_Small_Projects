## **Matrix Multiplication**

```python
b = torch.tensor([[6,7], [8,9], [10,11]])
a = torch.tensor([[5,6,5],[5,6,7]])
```

```python
# only possible if shape matchs correctly

# no of columns of a = no of rows of b

torch.mm(a,b)
```

```python
a@b
```

### **Dot Product**

```python
a = torch.tensor([1,2,3,4])
b = torch.tensor([5,6,7,8])
```

```python
a.dot(b)
```
