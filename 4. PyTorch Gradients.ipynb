# **PyTorch Gradients**
---
---

```python
import torch
```

```python

x = torch.tensor(2.0, requires_grad = True)
```

```python
# y is a function of x

y = 2*x**4 + x**3 + 3*x**2 + 5*x + 1
y
```

```python
type(y)
```

```python
y.backward()
```

```python
#dy/dx

y_dash = x.grad
y_dash
```

**Conclusion :** 93 is the slope of polynimial(dy/dx) when x = 2

### **Series of Equation - Gradient**

```python
x = torch.tensor([[1.,2.,3.],[3., 2., 1.]], requires_grad = True)
x
```

```python
y = 3*x + 2
y
```

```python
z = 2*y**2
z
```

```python
out = z.mean()
```

```python
out
```

```python
out.backward()
```

```python
x.grad
```
