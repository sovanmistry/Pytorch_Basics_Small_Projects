# **RANDOM VALUES**

```python
import torch
```

---
---
## **[3,4] Tensor with Uniformly Distributed Random Values**

lower Limit = 0, higher limit = 1

```python
torch.rand(3,4)
```

---
---
## **[3,4] Tensor with Normally Distributed Random Values**

Mean = 0, Standard Deviation = 1

```python
torch.randn(3,4)
```

---
---
## **[3,4] Tensor Random Integer Values between 0 to 10**

```python
torch.randint(low = 0, high = 11, size = (3,4))
```

---
---
## **[3,4] Tensor which uses shape of other tensor**

```python
other_tensor = torch.zeros(3,8)
```

```python
torch.rand_like(other_tensor)
```

```python
torch.randint_like( other_tensor, low = 0, high = 11 )
```
